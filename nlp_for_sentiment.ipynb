# NLP Algorithm

```python
#import packages
from gensim.summarization import keywords
import spacy
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stopwords = set(stopwords.words('english'))
from spacy.lang.en import English
nlp = English()
nlp.max_length = 10000000
import lyricsgenius
import pandas as pd
from textblob import TextBlob 
```

```python
#define Genius API authentication
genius = lyricsgenius.Genius('ERVmP0LvFIevOaSvPo9Vbt-HT9aPWPbQ_-FCU8Ryme5VrBOVbvxRPcnwPnNHqbJR')
```

```python
def get_lyrics(title, artist):
    ''' 
    Function to return lyrics of each song using Genius API
    '''
    try:
        return genius.search_song(title, artist).lyrics
    except:
        return 'not found'
    
def get_lyric_sentiment(lyrics): 
    ''' 
    Function to return sentiment score of each song
    '''
    analysis = TextBlob(lyrics) 
    return analysis.sentiment.polarity
```

```python
#Use get_lyrics funcion to get lyrics for every song in dataset
lyrics = songs.apply(lambda row: get_lyrics(row['Track Name'], row['Artist']), axis =1)
songs['Lyrics'] = lyrics
songs = songs.drop(songs[songs['Lyrics'] == 'not found'].index) #drop rows where lyrics are not found on Genius

#Use get_lyric_sentiment to get sentiment score for all the song lyrics
sentiment = songs.apply(lambda row: get_lyric_sentiment(row['Lyrics']), axis =1)
songs['Sentiment'] = sentiment
```
